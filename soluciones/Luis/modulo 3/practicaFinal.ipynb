```python
txtSinProcesar=''

while(len(txtSinProcesar)<=0):
    txtSinProcesar=input('Dame un textaco no vacío')
txt=txtSinProcesar.lower()

letrasNoTiene3letras=True
letras=['', '', '']

while(letrasNoTiene3letras):
    tresLetras=input('Dame tres letras seguidas')
    try:
        tresLetras=str(tresLetras).lower()
        if(len(tresLetras) != 3):
            print('Eso no son tres letras')
            continue
        for index, character in enumerate(tresLetras):
            if character.isalpha():
                letras[index]=tresLetras[index]
            else: 
                print('Mu mal, repite, he dicho tres Letras')
                raise ValueError
            
        letrasNoTiene3letras=False

    except ValueError:
        print('Eso no vale, repite')

pythonApareceEnElTexto = "Python".lower() in txt

nAparicionesLetra1 = txt.count(letras[0].lower())
nAparicionesLetra2 = txt.count(letras[1].lower())
nAparicionesLetra3 = txt.count(letras[2].lower())

print(f"Texto antes de procesarlo: {txtSinProcesar}")

print("\nResultados:")

print(f"\n  - El carácter '{letras[0]}' aparece {nAparicionesLetra1} veces")

print(f"\n  - El carácter '{letras[1]}' aparece {nAparicionesLetra2} veces")

print(f"\n  - El carácter '{letras[2]}' aparece {nAparicionesLetra3} veces")

longitudDelTexto=len(txtSinProcesar.split())

print(f"\n  - nPalabras: {len(txtSinProcesar.split())}")

primeraLetra=txtSinProcesar[0]

print(f"\n  - Primera letra: {primeraLetra}")

ultimaLetra=txtSinProcesar[-1]

print(f"\n  - Última letra: {ultimaLetra}")

textoAlReves=txtSinProcesar[::-1]

print(f"\n  - alverrés: {textoAlReves}")

pythonApareceEnElTexto = "Python".lower() in txt

print(f"\n  - ¿'Python' está presente en el texto?: \n {'Python'.lower() in txt} ")
```
